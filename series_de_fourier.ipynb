```python
import numpy as np
import matplotlib.pyplot as plt
```

Letra A

```python
N=13
n=np.arange(-20,20)
w0=2*np.pi/N
x=0

for k in range(-6,6):
  if k == -3 or k == 3:
    x=x+np.exp(1j*np.pi/(k*2))*np.exp(1j*k*w0*n)
    
plt.stem(n,x)
```

Letra C

```python
N=6
n=np.arange(-20,20)
w0=2*np.pi/N
x=0

for k in range(0,N):
  x=x+((1/N)*(2+2*np.cos(k*2*np.pi/3)+(-1)**k))*np.exp(1j*k*w0*n)

plt.stem(n,x)
```
